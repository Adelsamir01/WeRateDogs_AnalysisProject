# Cleaning the Data

## Cleaning archive_clean

```python
# first, creating new data frame from the original
archive_clean = twitter_archive.copy()
```

```python
archive_clean.info()
```

```python
# define: dropping all retweets

# code:
archive_clean = archive_clean[archive_clean['retweeted_status_id'].isnull()]
archive_clean = archive_clean[archive_clean['retweeted_status_user_id'].isnull()]
archive_clean = archive_clean[archive_clean['retweeted_status_timestamp'].isnull()]
```

```python
# test:
archive_clean.info()
```

```python
#test: no text containing (RT) 
archive_clean
```

```python
# define: dropping all unnessessary columns that miss so many values and unneeded ones like expanded_urls

# code:
archive_clean = archive_clean.drop(['in_reply_to_status_id', 
                                        'in_reply_to_user_id',
                                        'retweeted_status_id',
                                        'retweeted_status_user_id',
                                        'retweeted_status_timestamp',
                                         'expanded_urls'], axis=1)
```

```python
# test:
archive_clean.info()
```

```python
#define: fixing typo mistakes

#code:
archive_clean.loc[archive_clean['rating_numerator']==80, ['rating_numerator']] = 10
archive_clean.loc[archive_clean['rating_denominator']==80, ['rating_denominator']] = 10
archive_clean.loc[archive_clean['rating_numerator']==44, ['rating_numerator']] = 11
archive_clean.loc[archive_clean['rating_denominator']==40, ['rating_denominator']] = 10
```

```python
#test:
len(archive_clean[archive_clean['rating_numerator'] > 15])
```

```python
#define: cleaning irrelevant data like zeros or outliers

#code:
archive_clean = archive_clean[archive_clean['rating_numerator'] != 0 ]
archive_clean = archive_clean[archive_clean['rating_denominator'] >= 10 ]
archive_clean = archive_clean[archive_clean['rating_numerator'] <= 20 ]
archive_clean = archive_clean[archive_clean['rating_denominator'] < 20 ]
```

```python
#test:
archive_clean.info()
```

```python
# test:
len(archive_clean[archive_clean['rating_numerator'] == 0])
```

```python
# test:
len(archive_clean[archive_clean['rating_numerator'] > 20])
```

```python
# test:
len(archive_clean[archive_clean['rating_denominator'] < 10])
```

```python
# test:
len(archive_clean[archive_clean['rating_denominator'] > 20])
```

```python
# test:
archive_clean.info()
```

```python
# define: cleaning None values with NaN

#code:
archive_clean =  archive_clean.replace('None', np.nan)
```

```python
# test:
archive_clean.info()
```

```python
#define: creating new column to combine dog classifcations

#code:
archive_clean.loc[archive_clean['doggo'] == 'doggo', 'dog_class'] = 'doggo'
archive_clean.loc[archive_clean['floofer'] == 'floofer', 'dog_class'] = 'floofer'
archive_clean.loc[archive_clean['pupper'] == 'pupper', 'dog_class'] = 'pupper'
archive_clean.loc[archive_clean['puppo'] == 'puppo', 'dog_class'] = 'puppo'
```

```python
#test:
archive_clean.head(10)
```

```python
# define: droping the redundent columns 

# code:
archive_clean = archive_clean.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis = 1)
```

```python
# test:
archive_clean.info()
```

## Cleaning image.predictions

```python
images_clean = image_predictions.copy()
```

```python
images_clean.info()
```

```python
# define: renaming columns to be descriptive

# code:
images_clean= images_clean.rename(columns={'p1':'BreedFirstProbability',
                                               'p2':'BreedSecondProbability',
                                               'p3':'BreedThirdProbability'})
```

```python
# test
images_clean.info()
```

```python
# define: dropping all probabilities less than the half

# code:
images_clean = images_clean[images_clean['p1_conf'] >= 0.5]
```

```python
#test: 
images_clean.info()
```

```python
# define: selecting only the three needed columns to work with

# code:
images_clean = images_clean.filter(['tweet_id',
                                    'BreedFirstProbability',
                                    'p1_conf'] )
```

```python
#test
images_clean.info()
```

```python
# define: dropping the less frequent stuff in the data

#code
images_clean = images_clean.groupby('BreedFirstProbability').filter(lambda x: len(x) > 5)
```

```python
#test
images_clean.info()
```
